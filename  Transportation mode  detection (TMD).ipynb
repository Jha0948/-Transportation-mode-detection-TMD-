```python
import numpy as np
import pandas as pd
import time
```

```python
# 1
#Read the dataset “cleaned.csv”
dataset = pd.read_csv('cleaned.csv')
dataset.head()
```

```python
#a
# Output the number of rows, columns, and columns names of the input data. 
dataset.shape
```

```python
#b) Find out the number of unique users present in the dataset and output the value (Column “user”)
unq = dataset.drop_duplicates(['user'])
print(len(unq))
```

```python
#2 Determine the number of unique sequences
# a) Output using a table the number of unique sequences present for each transportation mode for
 # each user.
dataset['shift_user'] = dataset['user'].shift(-1)
dataset['shift_class'] = dataset['class'].shift(-1)
dataset['new_user'] = False
dataset.loc[dataset['shift_user']!=dataset['user'] , 'new_user'] =True
dataset['new_class'] = False
dataset.loc[dataset['shift_class']!=dataset['class'] ,'new_class'] = True
dataset
```

```python
# convert time to datetime
dataset['time'] = pd.to_datetime(dataset['timestamp'] )
dataset
```

```python
dataset['shift_time'] = dataset['time'].shift(-1)
dataset
```

```python
# make a new sequence of new  user , new class and new time 
dataset['timediff'] = (dataset['shift_time'] -dataset['time']).astype('timedelta64[ms]')
dataset['new_time'] = False
dataset.loc[dataset['timediff']>10000,'new_time'] = True

dataset['new_seq'] = dataset[['new_user' ,'new_class','new_time']].any(axis = 'columns')
dataset
```

```python
#calculate total no of sequence
dataset['new_id'] = 0
dataset.loc[dataset['new_seq'],'new_id'] =1

dataset['unq_id']= dataset['new_id'].cumsum()
dataset
```

```python
unq_seq = pd.DataFrame(dataset.groupby(['user','class'])['new_id'].sum())
unq_seq
```

```python
#b) Output the time taken for your code to run this particular section of code 

start = time.process_time()
dataset['shift_user'] = dataset['user'].shift(-1)
dataset['shift_class'] = dataset['class'].shift(-1)
dataset['new_user'] = False
dataset.loc[dataset['shift_user']!=dataset['user'] , 'new_user'] =True
dataset['new_class'] = False
dataset.loc[dataset['shift_class']!=dataset['class'] ,'new_class'] = True

dataset['time'] = pd.to_datetime(dataset['timestamp'] )
dataset['shift_time'] = dataset['time'].shift(-1)
dataset['timediff'] = (dataset['shift_time'] -dataset['time']).astype('timedelta64[ms]')
dataset['new_time'] = False
dataset.loc[dataset['timediff']>10000,'new_time'] = True

dataset['new_seq'] = dataset[['new_user' ,'new_class','new_time']].any(axis = 'columns')
dataset

dataset['new_id'] = 0
dataset.loc[dataset['new_seq'],'new_id'] =1

dataset['unq_id']= dataset['new_id'].cumsum()
dataset
unq_seq = pd.DataFrame(dataset.groupby(['user','class'])['new_id'].sum())
unq_seq  
print(time.process_time() - start)
```
